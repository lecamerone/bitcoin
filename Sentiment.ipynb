```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bitcoin.csv')
```

```python
tweet = np.zeros(len(df))
for i in df.index:
    classifier = TextBlob(df.Content.ix[i])
    tweet[i] = classifier.sentiment.polarity

```

```python
tweet
```

```python
np.min(tweet)
```

```python
plt.plot(tweet)
plt.show()
```

```python
sent = np.zeros(len(tweet) - 1)
for i in range(len(tweet) - 1):
    if i == 0:
        sent[i] = tweet[i]
    else:
        sent[i] = sent[i-1] + tweet[i]

plt.plot(sent)
plt.show()
```
