```python
import gdax
import pandas as pd
from datetime import datetime
import datetime
import numpy as np
import matplotlib.pyplot as plt

client = gdax.PublicClient()
x = client.get_product_historic_rates('LTC-USD', granularity=60*60*24)
```

```python
df = pd.DataFrame(x)
```

```python
for i in range(len(x)):
    df[0][i] = (
    datetime.datetime.fromtimestamp(
        int(df[0][i])
    ).strftime('%Y-%m-%d %H:%M:%S')
    )
```

```python
d = df.iloc[::-1]
d = d.reset_index(drop=True)
d
```

```python
close = d[4]
x = np.diff(np.log(close))
n = len(x)
```

```python
plt.plot(x)
plt.show()
```

```python
maxR = np.amax(x)
minR = np.amin(x)
support = np.linspace(minR, maxR, num = 50)
```

```python
count = np.zeros(50)
for i in range(len(x)):
    for j in range(len(support)):
        if x[i] <= support[j]:
            count[j] = count[j] + 1/n

count
```

```python
plt.plot(support,count)
plt.show()
```

```python
num = 50
Xsupport = np.linspace(minR, maxR, num = num)
count1 = np.zeros(num)
for i in range(len(x)):
    for j in range(len(Xsupport)-1):
        if x[i] < Xsupport[j+1] and x[i] >= Xsupport[j] :
            count1[j] = count1[j] + 1/n

count1
```

```python
plt.plot(Xsupport, count1)
plt.show()
```
