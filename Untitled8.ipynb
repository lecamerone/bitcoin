```python
import quandl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = quandl.get("BCHARTS/BITSTAMPUSD", returns="numpy",  start_date="2013-01-01", end_date="2017-09-26")
```

```python
d = pd.DataFrame(data)
d
```

```python
x = np.diff(np.log(d.Close))
n = len(x)
```

```python
plt.plot(x)
plt.show()
```

```python
x[x>1e308] = 0
x[np.isnan(x)] = 0
x[x<-1e308] = 0



maxR = np.amax(x)
minR = np.amin(x)
```

```python
support = np.linspace(minR, maxR, num = 50)
```

```python

count = np.zeros(50)
for i in range(len(x)):
    for j in range(len(support)):
        if x[i] <= support[j]:
            count[j] = count[j] + 1/n

count
            

        
```

```python
plt.plot(support,count)
plt.show()
```

```python
num = 50
Xsupport = np.linspace(minR, maxR, num = num)
count = np.zeros(num)
for i in range(len(x)):
    for j in range(len(Xsupport)-1):
        if x[i] < Xsupport[j+1] and x[i] >= Xsupport[j] :
            count[j] = count[j] + 1/n

count
```

```python
plt.plot(Xsupport, count)
plt.show()
```
